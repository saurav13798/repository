### 3. Aggeration operations in numpy

```python
# find out average marks inside the list
marks = [83, 72, 67, 89, 93, 70, 52, 35]
```

```python
# Initilize sum to 0
s = 0

# Apply for loop 
for i in marks:
    s = s + i

# find out number of student
count = len(marks)

#Average marks
avg = s/count

# Print average marks
print(f"Average of {marks} : {avg:.2f}")
```

```python
avg1 = np.mean(marks)
avg1
```
